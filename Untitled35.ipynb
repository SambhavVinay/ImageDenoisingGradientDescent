```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def tv_denoise(img, weight=0.1, lr=0.2, iterations=200):
    """
    Total Variation Denoising using gradient descent
    img: input noisy grayscale image
    weight: regularization strength
    lr: learning rate
    iterations: number of iterations
    """
    u = img.astype(np.float32)
    for i in range(iterations):
        # compute gradients
        u_x = np.roll(u, -1, axis=1) - u
        u_y = np.roll(u, -1, axis=0) - u

        # divergence
        div = (u_x - np.roll(u_x, 1, axis=1)) + (u_y - np.roll(u_y, 1, axis=0))

        # gradient descent step
        u = u + lr * (div - weight * (u - img))

        if (i+1) % 20 == 0:
            print(f"Iteration {i+1}/{iterations} complete")

    return np.clip(u, 0, 255).astype(np.uint8)

# Load your uploaded image
img_path = "/content/image_2025-10-28_205134795.png"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Denoise
denoised = tv_denoise(img, weight=0.2, lr=0.2, iterations=1200)

# Display
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Original Noisy")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Denoised")
plt.imshow(denoised, cmap='gray')
plt.axis('off')
plt.show()


```
